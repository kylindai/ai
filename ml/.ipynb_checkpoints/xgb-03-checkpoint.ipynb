```python
import pandas as pd
import numpy as np
import xgboost as xgb

%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt

from sklearn import metrics
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.datasets import load_iris

from xgboost import XGBClassifier
from xgboost import XGBRegressor
from xgboost import plot_importance

from sklearn2pmml import PMMLPipeline
from sklearn2pmml import sklearn2pmml 
```

```python
# 加载数据集
iris = load_iris()

# 取前100行特征
data = iris.data[:100]
print(data.shape)
print(data[:5])

# 取前100行标注
label = iris.target[:100]
print(label)
```

```python
# split training and test data
X_train, X_test, y_train, y_test = train_test_split(data, label, random_state=0)

# for xgb input
dtrain = xgb.DMatrix(X_train, label=y_train)
# dtest = xgb.DMatrix(X_test, label=y_test)
dtest = xgb.DMatrix(X_test)
```

```python
params = {
    'booster': 'gbtree',
    'objective': 'binary:logistic',
    'eval_metric': 'auc',
    'max_depth': 4,
    'lambda': 10,
    'subsample': 0.75,
    'colsample_bytree': 0.75,
    'min_child_weight': 2,
    'eta': 0.025,
    'seed': 0,
    'nthread': 8,
    'silent': 1
}

watchlist = [(dtrain, 'train')]

# 模型训练
bst = xgb.train(params, dtrain, num_boost_round=100, evals=watchlist)
```

```python
ypred = bst.predict(dtest)
# print(ypred)

#
# y_pred = (ypred >= 0.5) * 1
y_pred = np.where(ypred > 0.5, 1, 0)

print('AUC: %.4f' % metrics.roc_auc_score(y_test, ypred))
print('ACC: %.4f' % metrics.accuracy_score(y_test, y_pred))
print('Recall: %.4f' % metrics.recall_score(y_test, y_pred))
print('F1-Score: %.4f' % metrics.f1_score(y_test, y_pred))
print('Precesion: %.4f' % metrics.precision_score(y_test, y_pred))

metrics.confusion_matrix(y_test, y_pred)
```

```python
# 模型树加载

import re

booster_trees = {}

with open(model_file, 'r') as f:
    
    lines = f.readlines()
#     print(len(lines))
    
    for i in range(0, len(lines), 4):
        line = lines[i]
        if line.startswith('booster['):
            
            tree = {}
            
            # tree_no
            tree_no = re.search(r'\[(\d+)\]', line).group(1)
#             print('tree_no = {tree_no}'.format(tree_no=tree_no))

            # 根节点
            root_line = lines[i+1]
            # 根节点表达式
            root_exp_match = re.search(r'\[(.+)<(.+)\]', root_line)
#             print(root_exp_match.group(1) + ' < ' + root_exp_match.group(2))
            root = {
                'f': root_exp_match.group(1),
                'v': float(root_exp_match.group(2))
            }
#             print('root_exp = ' + str(root))
            tree['root'] = root
            
            # 左子树
            left_line = lines[i+2]
            left = re.search(r'=(.+)', left_line).group(1)
#             print('left = ' + left)
            tree['left'] = float(left)
            
            # 右子树
            right_line = lines[i+3]
            right = re.search(r'=(.+)', right_line).group(1)
#             print('right = ' + right)
            tree['right'] = float(right)
            
            booster_trees['booster_' + tree_no] = tree
    
booster_trees
```

```python
# 样本带入模型
def get_feature_score(tree, f_name, f_value):
    
    score = 0
    
    f = tree['root']['f']
    v = tree['root']['v']

    if f_name == f:
        if f_value < v:
            score = tree['left']
            print('    %s [yes] %.10f < %.10f' % (f_name, f_value, v))
        elif f_value >= v:
            score = tree['right']
            print('    %s [no] %.10f >= %.10f' % (f_name, f_value, v))
        else:
            score = tree['left']
            print('    %s [mission] %.10f < %.10f' % (f_name, f_value, v))
        
    return score

# 第i个样本在模型中的分数
def get_pred_score(i):
    
    # 样本1在模型中的分数
    f1 = X_test[i,0]
    f2 = X_test[i,1]
    f3 = X_test[i,2]
    f4 = X_test[i,3]

    pred_score = 0

    for booster,tree in booster_trees.items():
        score = 0
        print('>>> %s: score = %.10f' % (booster, pred_score))
        score += get_feature_score(tree, 'f1', f1)
        score += get_feature_score(tree, 'f2', f2)
        score += get_feature_score(tree, 'f3', f3)
        score += get_feature_score(tree, 'f4', f4)
        print('    %s: score   %.10f' % (booster, score))
        pred_score += score        
        print('<<< %s: score = %.10f\n' % (booster, pred_score))
    
    return pred_score
```
