```python
from scipy import sparse
import numpy as np

row = np.array([0, 3, 1, 0])
col = np.array([0, 3, 1, 2])
data = np.array([6, 5, 7, 8])

# Coordinate (COO)
# 生成矩阵
# data[i] = X[row][col]
# data[0] = 6，生成的矩阵 X[0][0] = 6
sparse.coo_matrix((data, (row, col)), shape=(4,4)).toarray()
```

```python
from scipy import sparse
import numpy as np

indptr = np.array([0, 2, 3, 6])
indices = np.array([0, 2, 2, 0, 1, 2])
data = np.array([1, 2, 3, 4, 5, 6])

# Compressed Sparse Column (CSC)
# 生成矩阵
# indptr:  代表矩阵中每一列所存的数在data中的开始和结束索引（左闭右开）
#          上例indptr有3列：[0,2) [2,3) [3,6)
#          表示data[0,2)即 1,2 放在第 1 列，data[2,3)即 3 放在第 2 列           
# indices: 代表data[i]中的数据位于矩阵的行数
#          上例indices[0]表示data[0]即 1 放在第 0 行
#          上例indices[1]表示data[1]即 2 放在第 2 行 
sparse.csc_matrix((data, indices, indptr), shape=(3, 3)).todense()
```

```python
from scipy import sparse
import numpy as np

indptr = np.array([0, 2, 3, 6])
indices = np.array([0, 2, 2, 0, 1, 2])
data = np.array([1, 2, 3, 4, 5, 6])

# Compressed Sparse Row (CSR)
# 生成矩阵
# indptr:  代表矩阵中每一行所存的数在data中的开始和结束索引（左闭右开）
#          上例indptr有3行：[0,2) [2,3) [3,6)
#          表示data[0,2)即 1,2 放在第 1 行，data[2,3)即 3 放在第 2 行
# indices: 代表data[i]中的数据放在矩阵的列数
#          上例indices[0]表示data[0]即 1 放在第 0 列
#          上例indices[1]表示data[1]即 2 放在第 2 列 
sparse.csr_matrix((data, indices, indptr), shape=(3, 3)).todense()
```
