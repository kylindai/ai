```python
import numpy as np
import tensorflow as tf
```

```python
a = np.arange(8).reshape(2, 4)
b = np.arange(8, 16).reshape(2, 4)
c = np.arange(12, 20).reshape(2, 4)
print("a", a)
print("b", b)
print("c", c)
```

```python
a = tf.Variable(a, dtype=tf.float32)
b = tf.Variable(b, dtype=tf.float32)
c = tf.Variable(c, dtype=tf.float32)

idx = tf.SparseTensor(
    indices=[[0,0],[0,2],[1,0],[1,1]],
    values=[1,2,2,0], dense_shape=(2,3))

result = tf.nn.embedding_lookup_sparse([a,b,c], idx, None,
                                    partition_strategy='mod',
                                    combiner='sum')
init = tf.global_variables_initializer()

with tf.Session() as sess:
    sess.run(init)
    res = sess.run(result)
    print('result:\n', res)
```
